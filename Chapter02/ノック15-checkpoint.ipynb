```python
# 商品毎の平均金額を算出し、欠損値に代入する

# まずは商品Aの列のみを抽出

uriage_data[uriage_data["item_name"] == "商品A"]
```

```python
# item_priceの欠損地をすべて100に変換し、元データに代入

uriage_data.loc[uriage_data["item_name"] == "商品A","item_price"] = uriage_data.loc[uriage_data["item_name"] == "商品A","item_price"].fillna(100)
uriage_data.loc[uriage_data["item_name"] == "商品A","item_price"]
```

```python
# for文を用いて他の値も書き換える

item_list = [np.arange(65,91)]
price_list = [np.arange(100,2601,100)]

for (a,b) in zip(item_list,price_list):
    uriage_data.loc[uriage_data["item_name"] == "商品" + chr(a),"item_price"] = uriage_data.loc[uriage_data["item_name"] == "商品" + chr(a),"item_price"].fillna(b)
```
